# NYC TLC Taxi Trip Data Analysis

## Background
The New York City Taxi and Limousine Commission (NYC TLC) is a regulatory agency responsible for overseeing the taxi, for-hire vehicle, and paratransit industries in New York City. Established in 1937, NYC TLC ensures the safety, quality, and accessibility of these services while promoting fair competition and protecting the rights of both passengers and drivers. The business process for NYC TLC involves the regulation and monitoring of taxi and for-hire vehicle services. This includes issuing permits, conducting inspections, enforcing regulations, and collecting data on taxi trips to analyze trends and improve service quality. Tipping is a common practice in the taxi industry where passengers provide additional gratuity to drivers for their service. Understanding tipping behavior is crucial for NYC TLC to ensure fair compensation for drivers and to improve overall customer satisfaction.

## Gap Analysis
Despite the wealth of data available from the NYC TLC, there is a notable gap in understanding customer tipping behavior within the taxi service. While the dataset includes detailed information on trip fares, distances, and payment types, it lacks comprehensive insights into the factors influencing tipping patterns. For instance, variables such as trip duration, time of day, and passenger demographics are not fully explored, which limits the ability to identify trends or anomalies in tipping behavior. Additionally, the dataset does not provide contextual information about the service quality or customer satisfaction, which are critical in understanding why some passengers tip generously while others do not.

## Problem Statement
The primary issue at hand is the lack of clarity regarding the factors that influence tipping behavior among taxi passengers in New York City. Given the available dataset, it is essential to analyze how variables such as trip distance, fare amount, and time of day correlate with the amount tipped by passengers. Understanding these relationships will help NYC TLC identify patterns in customer behavior, which can inform strategies to enhance service quality and potentially increase tipping rates. Furthermore, insights derived from this analysis could lead to targeted training for drivers on customer service practices that encourage tipping, ultimately benefiting both drivers and the overall taxi service experience.

## Data Understanding

### 1. VendorID
The VendorID column indicates the taxi company providing the service. It is a categorical variable with values representing different taxi companies.

### 2. lpep_pickup_datetime
The lpep_pickup_datetime column records the date and time when the taxi trip began. It is a datetime variable that can be used to analyze temporal patterns in taxi trips.

### 3. lpep_dropoff_datetime
The lpep_dropoff_datetime column records the date and time when the taxi trip ended. It is a datetime variable that can be used to calculate trip duration and analyze temporal patterns in taxi trips.

### 4. store_and_fwd_flag
The store_and_fwd_flag column indicates whether the trip record was held in the vehicle before being sent to the vendor. It is a categorical variable with values 'Y' (Yes) or 'N' (No).

### 5. RatecodeID
The RatecodeID column indicates the rate code in effect at the beginning of the trip. It is a categorical variable with values representing different rate codes.

### 6. PULocationID
The PULocationID column indicates the ID of the pickup location. It is a numerical variable that can be used to identify the pickup zone.

### 7. DOLocationID
The DOLocationID column indicates the ID of the dropoff location. It is a numerical variable that can be used to identify the dropoff zone.

### 8. passenger_count
The passenger_count column indicates the number of passengers in the taxi. It is a numerical variable that can be used to analyze passenger load.

### 9. trip_distance
The trip_distance column indicates the distance traveled during the trip in miles. It is a numerical variable that can be used to analyze trip lengths.

### 10. fare_amount
The fare_amount column indicates the base fare amount for the trip. It is a numerical variable that can be used to analyze fare structures.

### 11. extra
The extra column indicates additional charges for the trip. It is a numerical variable that can be used to analyze additional fees.

### 12. mta_tax
The mta_tax column indicates the MTA tax for the trip. It is a numerical variable that can be used to analyze tax structures.

### 13. tip_amount
The tip_amount column indicates the tip amount provided by the passenger. It is a numerical variable that can be used to analyze tipping behavior.

### 14. tolls_amount
The tolls_amount column indicates the tolls paid during the trip. It is a numerical variable that can be used to analyze toll expenses.

### 15. ehail_fee
The ehail_fee column indicates the fee for electronic hail requests. It is a numerical variable that can be used to analyze electronic hail usage.

### 16. improvement_surcharge
The improvement_surcharge column indicates the improvement surcharge for the trip. It is a numerical variable that can be used to analyze surcharge structures.

### 17. total_amount
The total_amount column indicates the total amount paid by the passenger. It is a numerical variable that can be used to analyze overall trip costs.

### 18. payment_type
The payment_type column indicates the method of payment used by the passenger. It is a categorical variable with values representing different payment methods.

### 19. trip_type
The trip_type column indicates the type of trip. It is a categorical variable with values representing different trip types.

### 20. congestion_surcharge
The congestion_surcharge column indicates the congestion surcharge for the trip. It is a numerical variable that can be used to analyze congestion surcharge structures.

# Data Wrangling

Data wrangling is a critical process in data science that involves the cleaning, structuring, and enriching of raw data to make it suitable for analysis in various applications, including machine learning and data analytics. Also known as data munging or data preparation, this process addresses common data quality issues such as missing values, duplicates, outliers, and formatting inconsistencies. The ultimate goal is to transform unstructured or problematic data into clean datasets that can be effectively analyzed, supporting informed decision-making.

The importance of data wrangling lies in its ability to ensure that the resulting datasets are clean, consistent, and reliable, which is essential for tasks such as building machine learning models, performing data analytics, creating data visualizations, generating business intelligence reports, and making informed executive decisions. Without proper data wrangling, the results of data analysis can be misleading, leading to inaccurate conclusions and flawed business decisions.

## The Data Wrangling Process

The data wrangling process is essential for preparing raw data for analysis and typically involves several key steps. Each step is designed to address specific challenges associated with data quality and usability. Activities within these steps include discovering data sources, structuring data into a unified format, cleaning inconsistencies, enriching datasets with additional information, and validating the accuracy of the transformed data. By systematically executing these activities, data scientists can ensure that the final datasets are clean, reliable, and ready for insightful analysis. Ultimately, this process enhances the overall quality and effectiveness of data-driven decision-making.

### Discovering

The discovering phase focuses on the assessment of the quality of the complete dataset, including its sources and formats. This initial evaluation is crucial for identifying potential issues that could affect the analysis. During this phase, the quality of the data set is assessed, and data sources and formats are identified. Quality issues such as missing data, formatting inconsistencies, errors, or bias, and outliers that might skew the analysis are highlighted and addressed. The findings are typically documented in a data quality report or a more technical document known as a data profiling report, which includes statistics, distributions, and other results.

```python
real_trips = pd.read_csv('NYC TLC Trip Record.csv')
trips = real_trips.copy()
```

#### Identifying Data Sources and Formats: 
Data sources and formats are identified to understand where the data is coming from and how it is structured.

```python
trips.info()
```

```python
# Create dictionary of unique values
trips_unique = dict()
for column in trips.columns:
    trips_unique[column] = list(trips[column].unique())
trips_unique

# Check for unique values
trips_nunique = trips.nunique().reset_index().sort_values(0, ascending=True).rename(columns={0: 'Unique Value Counts', 'index': 'Column Name'})
trips_nunique = trips_nunique[trips_nunique['Unique Value Counts'] > 1]
trips_nunique['Unique Value Lists'] = trips_nunique['Column Name'].map(trips_unique)
trips_nunique
```

#### Highlighting Quality Issues: 
Quality issues such as missing data, formatting inconsistencies, errors, or bias, and outliers that might skew the analysis are highlighted.

```python
# Check for missing values
trips_isnull_sum = trips.isnull().sum().reset_index()
trips_isnull_sum[1] = trips_isnull_sum[0] / len(trips) * 100

rename_columns = {0: 'Missing Value Counts', 1: 'Percentage (%)', 'index': 'Column Name'}
trips_isnull_sum = trips_isnull_sum.rename(columns=rename_columns)
trips_isnull_sum[trips_isnull_sum['Missing Value Counts']!=0]
```

```python
sample_values = dict()
for column in trips.columns:
    sample_values[column] = trips[column].head(1).values[0]
sample_values

# Check formatting inconsistencies
trips_dtypes = trips.dtypes.reset_index().rename(columns={0: 'Type', 'index': 'Column Name'})
trips_dtypes['Sample Value'] = trips_dtypes['Column Name'].map(sample_values)
trips_dtypes
```

# Exploratory Data Analysis (EDA)
Exploratory Data Analysis (EDA) is a fundamental process in data science that involves examining datasets to uncover underlying patterns, trends, and anomalies without making prior assumptions about the data. It is akin to exploring a new environment, where data scientists meticulously inspect various aspects of the dataset, such as distributions, relationships, and potential outliers. EDA employs a combination of statistical techniques and visualization tools to summarize and visualize data. EDA helps in understanding the data structure, identifying missing values, and detecting outliers, which are crucial for building accurate models.

EDA is essential for ensuring data quality and integrity before proceeding to more complex analyses. It helps in generating accurate models by identifying and addressing issues in the data. EDA prevents the use of wrong data and inefficient resource use by ensuring proper data preparation. It aids in creating the right types of variables, which is critical for effective data preparation.

## Univariate Analysis: 
Examines individual variables to understand their distributions and summary statistics. Suitable visualizations include histograms, box plots, and violin plots.

## Bivariate Analysis: 
Explores the relationship between two variables. Suitable visualizations include scatter plots, pair plots, and heatmaps.

## Multivariate Analysis: 
Analyzes interactions among multiple variables. Suitable visualizations include contour plots and 3D scatter plots.

## Visualization Techniques: 
Utilizes various charts to depict data distributions and trends. Common visualizations include bar charts, line charts, scatter plots, and heatmaps.

## Outlier Detection: 
Identifies anomalies within the data. Suitable tools include box plots, z-score analysis, and scatter plots.

## Statistical Tests: 
Validates hypotheses or discerns significant differences between groups. Common tests include t-tests, chi-square tests, and ANOVA.

## Cumulative Distribution Function (CDF) Plots: 
Quantify the probability that a variable will take a value less than or equal to a specific value, providing a cumulative measure of the distribution.

## Pair Plot: 
Visualizes pairwise relationships between variables in a dataset, displaying both the distributions of individual variables and their relationships.

## Joint Plot: 
Provides a detailed view of the relationship between two variables along with their individual distributions.

## Heatmap: 
Visualizes the correlation between different variables using color coding to represent the strength of the correlations.

## Contour Plot: 
Represents a 3-dimensional surface by plotting constant z slices, called contours, in a 2-dimensional format, allowing visualization of complex relationships between three variables.

## 3D Scatter Plot: 
Extends traditional scatter plots into three dimensions, allowing visualization of the relationship among three variables.
